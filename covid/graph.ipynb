```python
with open('StatusOverTime.txt') as input_data:
        # read the lines and build list of persons of various categories
    lines=input_data.readlines()
    S= [int(line.strip().split()[0])for line in lines] 
    V= [int(line.strip().split()[1])for line in lines]
    I= [int(line.strip().split()[2])for line in lines]
    R= [int(line.strip().split()[3])for line in lines]
    D= [int(line.strip().split()[4])for line in lines]
       

print(S)
print(V)
print(I)
print(R)
print(D)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Plottting time steps vs S,V, I, R and D
fig, ax = plt.subplots(figsize=(8.5, 6), dpi=130)



ax.plot(range(len(S)),S, 'bo', label='susceptible')
ax.plot(range(len(V)),V, 'yo', label='vaccinated')  
ax.plot(range(len(I)),I, 'ro', label='infected')           
ax.plot(range(len(R)),R, 'go', label='recovered')
ax.plot(range(len(D)),D, 'ko', label='dead')
ax.set_title('Susceptible,Vaccinated, Infected, Recovered and Dead')
ax.set_xlabel('Day No')
ax.set_ylabel('Number of Susceptible/Vaccinated/Infected/Recovered/ Dead')

ax.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Plottting time steps vs S, I and R and D
fig, ax = plt.subplots(figsize=(8.5, 6), dpi=130)



ax.plot(S)
ax.scatter(range(len(S)), S, c='blue', s=13, label = 'Susceptible(blue)' )
ax.plot(V)
ax.scatter(range(len(V)), V,c='yellow', s=13, label = 'Vaccinated(Yellow)') 
ax.plot(I)
ax.scatter(range(len(I)), I,c='red', s=13, label = 'Infected(red)')           
ax.plot(R)
ax.scatter(range(len(R)), R,c='green', s=13, label = 'Recovered(green)')
ax.plot(D)
ax.scatter(range(len(D)), D, c='black',s=13, label = 'Dead(black)')
ax.set_title('Susceptible, vaccinated, Infected, Recovered and Dead')
ax.set_xlabel('Day No')
ax.set_ylabel('No of Susceptible/Vaccinated/Infected/Recovered/Dead')

ax.legend()

plt.show()
```
