```python
with open('SummaryStatusOverTime.txt') as input_data:
        # read the lines and build list of persons of various categories
    lines=input_data.readlines()
    S= [float(line.strip().split()[0])for line in lines] 
    I= [float(line.strip().split()[1])for line in lines]
    R= [float(line.strip().split()[2])for line in lines]
   
       

print(S)
print(I)
print(R)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# Plottting time steps vs S, I and R 
fig, ax = plt.subplots(figsize=(8.5, 6), dpi=130)



ax.plot(range(len(S)),S, 'bo', label='susceptible')
ax.plot(range(len(I)),I, 'ro', label='infected')           
ax.plot(range(len(R)),R, 'go', label='recovered')

ax.set_title('Susceptible, Infected, Recovered')
ax.set_xlabel('Day No')
ax.set_ylabel('Number of Susceptible/Infected/Recovered')

ax.legend()

plt.show()
```
